```
df = pd.read_csv("/content/players_20.csv")
df.head()
```

The Pandas documentation [here](https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html#applying-multiple-functions-at-once) should be a useful reference.

```
nationality_df = df.groupby(['international_reputation', 'nationality']).agg(player_avg_age=('age', 'mean'), player_count=('sofifa_id', 'count')).reset_index()
# nationality_df.columns = nationality_df.droplevel(0)
# nationality_df.columns = ['nationality', 'player_avg_age', 'player_count']
nationality_df.head(5)
```

```
nationality_df.dtypes
```

```
player_age_bar_chart = alt.Chart(nationality_df).mark_bar().encode(x=alt.X('nationality:O', title='Nationality'), 
                                                                   y=alt.Y('player_avg_age:Q', title='Avg Player Age')
                                                                  ).interactive()

player_age_bar_text = player_age_bar_chart.mark_text(
                                                    align='left',
                                                    baseline='middle',
                                                    dx=3  # Nudges text to right so it doesn't appear on top of the bar
                                                    ).encode(text='player_avg_age:Q')


player_age_bar_chart.properties(width=1600, height=400)
# nationality_df['player_avg_age'].plot.bar(figsize=(27, 9))
```

```
player_count_bar_chart = alt.Chart(nationality_df).mark_bar().encode(x=alt.X('nationality:O', title='Nationality'), 
                                                                     y=alt.Y('player_count:Q', title='# of Players')
                                                                  ).interactive()

player_count_bar_text = player_count_bar_chart.mark_text(
                                      align='left',
                                      baseline='middle',
                                      dx=3  # Nudges text to right so it doesn't appear on top of the bar
                                      ).encode(text='player_count:Q')


player_count_bar_chart.properties(width=1600, height=400)

# nationality_df['player_count'].plot.bar(figsize=(27, 9))
```

# Big Scatter for Week 4

```
player_count_bar_chart = alt.Chart(nationality_df).mark_circle(size=60).encode(x=alt.X('nationality:O', title='Nationality'), 
                                                                     y=alt.Y('player_count:Q', title='# of Players'),
                                                                     color='international_reputation'
                                                                  ).interactive()

player_count_bar_text = player_count_bar_chart.mark_text(
                                      align='left',
                                      baseline='middle',
                                      dx=3  # Nudges text to right so it doesn't appear on top of the bar
                                      ).encode(text='player_count:Q')


player_count_bar_chart.properties(width=1600, height=400)
```
